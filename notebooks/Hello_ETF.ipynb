```python
import yfinance as yf, pandas as pd, matplotlib.pyplot as plt

tickers = ["SPY", "QQQ"]
data = yf.download(tickers, start="2020-01-01")["Close"]
returns = data.pct_change().dropna()
cumulative = (1 + returns).cumprod()

cumulative.plot(title="Total Return 2020â€“2025")
plt.show()
```
